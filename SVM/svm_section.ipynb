```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dict = load_breast_cancer()
```

```python
cancer_dict.keys()
```

```python
print(cancer_dict.DESCR)
```

```python
cancer_dict.feature_names
```

```python
cancer_feature_df = pd.DataFrame(cancer_dict.data, columns=cancer_dict.feature_names)
```

```python
cancer_feature_df.head()
```

```python
from sklearn import preprocessing

x = cancer_feature_df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
cancer_feature_df = pd.DataFrame(x_scaled)
cancer_feature_df
```

```python
cancer_feature_df['mean perimeter'] = cancer_feature_df['mean perimeter']/cancer_feature_df['mean perimeter'].max()
```

```python
cancer_feature_df.head()
```

```python
cancer_feature_df['mean area'] = cancer_feature_df['mean area']/cancer_feature_df['mean area'].max()
```

```python
cancer_feature_df['mean radius'] = cancer_feature_df['mean radius']/cancer_feature_df['mean radius'].max()
```

```python
cancer_feature_df['mean texture'] = cancer_feature_df['mean texture']/cancer_feature_df['mean texture'].max()
```

```python
cancer_feature_df['worst radius'] = cancer_feature_df['worst radius']/cancer_feature_df['worst radius'].max()
```

```python
cancer_feature_df['worst texture'] = cancer_feature_df['worst texture']/cancer_feature_df['worst texture'].max()
```

```python
cancer_feature_df['worst perimeter'] = cancer_feature_df['worst perimeter']/cancer_feature_df['worst perimeter'].max()
```

```python
cancer_feature_df['worst area'] = cancer_feature_df['worst area']/cancer_feature_df['worst area'].max()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = cancer_feature_df
y = cancer_dict.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svm_classifier = SVC(kernel='poly', degree=5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search_dic = {'C': [0.01, 0.1, 1 , 8, 10 , 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]}
```

```python
grid_search = GridSearchCV(svm_classifier, search_dic)
```

```python
# grid_search.fit(X_train, y_train)
svm_classifier.fit(X_train, y_train)
```

```python
svm_classifier.intercept_
```

```python
grid_search.best_params_
```

```python
# prediction = grid_search.predict(X_test)
prediction = svm_classifier.predict(X_test) 
```

```python
prediction
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, prediction))
```

```python
print(classification_report(y_test, prediction))
```
