```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dict = load_breast_cancer()
```

```python
cancer_dict.keys()
```

```python
print(cancer_dict.DESCR)
```

```python
cancer_dict.feature_names
```

```python
cancer_feature_df = pd.DataFrame(cancer_dict.data, columns=cancer_dict.feature_names)
```

```python
cancer_feature_df.head()
```

```python
from sklearn import preprocessing
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

svm_classifier = SVC(kernel='rbf')

# x = cancer_feature_df.values #returns a numpy array
# min_max_scaler = preprocessing.MinMaxScaler()
# x_scaled = min_max_scaler.fit_transform(x)
# cancer_feature_df = pd.DataFrame(x_scaled)
# cancer_feature_df

pipe_steps = [('scaler', preprocessing.StandardScaler()), ('svc', svm_classifier)]
pipe = Pipeline(steps=pipe_steps)
```

###### ----------------------------------------------------------

```python
from sklearn.model_selection import train_test_split
```

```python
X = cancer_feature_df
y = cancer_dict.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
search_hyper_params = {'svc__C': [0.01, 0.1, 1 , 8, 10 , 100, 1000], 'svc__gamma': [2, 1, 0.1, 0.01, 0.001,0.0005, 0.0001]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(pipe, search_hyper_params, cv=25, n_jobs=1)
```

```python
# grid_search.fit(X_train, y_train)
grid_search.fit(X_train, y_train)
```

```python
pd.DataFrame(grid_search.cv_results_)
```

```python
svm_classifier.intercept_
```

```python
grid_search.best_params_
```

```python
prediction = grid_search.predict(X_test)
# prediction = svm_classifier.predict(X_test) 
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, prediction))
```

```python
print(classification_report(y_test, prediction))
```
